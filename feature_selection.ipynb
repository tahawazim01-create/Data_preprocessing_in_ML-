```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = pd.DataFrame({
    'Feature1': [1, 2, 3, 4],
    'Feature2': [2, 4, 6, 8],
    'Target': [0, 1, 0, 1]
})

# Correlation matrix
corr = df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.show()
```
