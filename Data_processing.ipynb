```python
import numpy as np
import pandas as pd

df = pd.DataFrame ({'Name' : ['Alice', 'Bob', 'charlie', 'Alice'],
                    'Age' : [25, np.nan, 35, 25 ],
                    'Salary': [50000, 60000, np.nan, 50000]})

# print(df)
df = df.drop_duplicates()

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)


print(df)
```
