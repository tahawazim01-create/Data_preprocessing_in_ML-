```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Dataset
df = pd.DataFrame({
    'Age': [20, 30, 40, 50],
    'Income': [20000, 40000, 60000, 80000]
})

# Min-Max Scaling
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df[['Age', 'Income']])
scaled_data
df_scaled = pd.DataFrame(
    scaled_data,
    columns=['Age_Scaled', 'Income_Scaled']
)

print(df_scaled)
```
